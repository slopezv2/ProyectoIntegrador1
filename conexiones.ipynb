```
configs = {"fs.azure.account.auth.type": "OAuth",
           "fs.azure.account.oauth.provider.type": "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider",
           "fs.azure.account.oauth2.client.id": dbutils.secrets.get(scope="secretosvictoriasecret",key="clientid"),
           "fs.azure.account.oauth2.client.secret": dbutils.secrets.get(scope="secretosvictoriasecret",key="secretid"),
           "fs.azure.account.oauth2.client.endpoint": f"https://login.microsoftonline.com/{dbutils.secrets.get(scope='secretosvictoriasecret',key='tenantid')}/oauth2/token"}
```

```
def mount_point(punto_montaje):
  dbutils.fs.mount(
  source = f"abfss://{punto_montaje}@almacenamientosss.dfs.core.windows.net/",
  mount_point = f"/mnt/{punto_montaje}",
  extra_configs = configs)
```

```
mount_point("bronce1")
```

```
mount_point("plata")
```

```
mount_point("oro")
```

```
mount_point("sandbox")
```
